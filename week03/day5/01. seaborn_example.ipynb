- 이를 기반으로하는 `seaborn`을 이번 실습에서 사용

```python
import seaborn as sns
```

## 꺽은선 그래프
- Line Plot
- 두 변수의 값에 따른 추이를 선으로 이은 그래프

```python
x = [1, 2, 3, 4]
y = [0.7, 0.2, 0.3, 1.4]

sns.lineplot(x=x, y=y)
```

## 막대 그래프
- Bar Plot
- 범주형 데이터의 값과 그 크기를 직사각형으로 표현

```python
x = [1, 2, 3, 4]        ## 범주형 데이터이기 때문에, 문자열이 들어가도 된다.
y = [0.4, 0.2, 0.1, 0.3]

sns.barplot(x=x, y=y)
```

## Plot의 속성
- 그림의 크기나 제목과 같은 속성들을 설정할 수 있다.
- matplotlib의 pyplot을 이용하여 쉽게 속성을 변경할 수 있다.

```python
import matplotlib.pyplot as plt
```

### 제목 추가하기

```python
x = ["A", "B", "C", "D"]
y = [0.4, 0.2, 0.1, 0.3]

sns.barplot(x=x, y=y)
## 제목 설정
plt.title("Bar Plot")

## 축 설명 추가
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

### 축의 범위 변경


```python
x = [1, 2, 3, 4]
y = [0.7, 0.2, 0.3, 1.4]

sns.lineplot(x=x, y=y)
plt.ylim(0, 10)
```

### 그래프의 크기 지정

```python
x = [1, 2, 3, 4]
y = [0.7, 0.2, 0.3, 1.4]

plt.figure(figsize=(50, 10)) ## (x, y)
sns.lineplot(x=x, y=y)
```
