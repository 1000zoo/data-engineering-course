## 라이브러리 불러오기

```python
import requests
import json
```

## 모든 DAG 불러오기

```python
username = "airflow"
password = "airflow"
url = "http://localhost:8080/api/v1/dags"

auth = requests.auth.HTTPBasicAuth(username, password)

res = requests.get(url, auth=auth)
dags_origin = res.text
print(dags_origin)
print(type(dags_origin))
```

```python
dags_results = json.loads(dags_origin)
print(dags_results)

dags = dags_results["dags"]
print(dags)
print(type(dags))
```

```python
for dag in dags:
    if dag["is_active"] and not dag["is_paused"]:
        print(dag["dag_id"])
```
